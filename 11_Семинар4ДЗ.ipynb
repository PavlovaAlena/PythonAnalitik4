```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =  pd.read_csv('/content/Analitik/kc_house_data.csv', sep=',')
df.head()
```

### 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(8, 5))
plt.hist(df['price'])
plt.xlabel('Количество')
plt.ylabel('Стоимость')
plt.title('Распределение стоимости недвижимости');
```

### 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['sqft_living'], bins=20)
plt.xlabel('Жилые кв. метры')
plt.ylabel('Количество')
plt.title('Распределение квадратуры жилой');
```

### 1.4 Изучите распределение года постройки

```python
x = df['yr_built'].value_counts()
print(x)
plt.figure(figsize=(8,5))
plt.bar(x.index, x.values);
```

## Задача 3
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corrDF = df.corr()
corrDF = np.round(corrDF, 1)
corrDF[np.abs(corrDF) < 0.3] = 0
corrDF
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corrDF, annot = True, linewidths=0.5, cmap='coolwarm');
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Вид')
plt.title('Распределение цены дома от оценки вида');
print("Наибольшую цену имеют дома с оценкой вида 4")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Количество этажей')
plt.title('Распределение цены дома от этажности');
print("Наибольшую цену имеют дома высотой в 2.5 этажа")
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('Цена')
plt.ylabel('Качество')
plt.title('Распределение цены дома от качества конструкции и дизайна');
print("Наибольшую стоимость имеют дома с оценкой качества конструкции и дизайна 13")
```

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg');
print("Чем больше у дома этажей над уровнем земли, тем выше его стоимость");
```

```python
sns.jointplot(x=df['lat'], y=df['long'], kind='kde');
print("Скученность показывает места расположения продаваемых домов");
```
